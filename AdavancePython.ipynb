ATM PROGRAM WITH FUNCTION

```python
def login(pin, correct_pin=1674):
    return pin == correct_pin


def deposit(balance, transactions):
    amt = int(input("Enter the amount for deposit : "))
    balance += amt
    transactions.append(f"Deposit of {amt}")
    print(f"Deposit Successful of {amt}")
    return balance


def withdraw(balance, transactions):
    amt = int(input("Enter the amount you want to withdraw : "))

    if amt > balance:
        print("Insufficient Balance")
    else:
        balance -= amt
        transactions.append(f"Withdraw of {amt}")
        print(f"Withdraw Successful of {amt}")

    return balance


def history(transactions):
    if not transactions:
        print("No Transactions")
    else:
        for t in transactions:
            print(t)


def view_balance(balance):
    print(f"Account balance : {balance}")

balance = 0
transactions = []
approve = False

for cnt in range(3):
    print(f"Available chances : {3 - cnt}")
    code = int(input("Enter your pin : "))

    if login(code):
        approve = True
        break
    else:
        print("Wrong Pin")

while approve:
    print("\n1 for Deposit")
    print("2 for Withdraw")
    print("3 for Balance Inquiry")
    print("4 for All Transactions")
    print("5 for Exit\n")

    ch = int(input("Enter Your Choice : "))

    if ch == 1:
        balance = deposit(balance, transactions)

    elif ch == 2:
        balance = withdraw(balance, transactions)

    elif ch == 3:
        view_balance(balance)

    elif ch == 4:
        history(transactions)

    elif ch == 5:
        print("Exiting...!")
        break

    else:
        print("Invalid Choice !!")

print("Thank you for using")
```

```python
parkLots = {i:{"alloted":False,"OName":None,"Vno":None,"ph":None,"rs":5} for i in range(1,11)}
veh = set()
exits = {}
totalAmt = 0
his = 1
def checkEmpty(pNo):
  if (parkLots[pNo]["alloted"]):
    return False
  else:
    return True

def makeAllot(pNo,name,vno,ph):
  parkLots[pNo]["OName"] = name
  parkLots[pNo]["Vno"] = vno
  parkLots[pNo]["ph"] = ph
  parkLots[pNo]['alloted'] = True
  print(f"Vehicle no {vno} alloted to slot {pNo} successfully")

def makeCheckOut(pNo):
  global totalAmt
  if checkEmpty(pNo):
    print("Already Empty !!")
    return
  exits[his] = parkLots[pNo]
  veh.remove(parkLots[pNo]['Vno'])
  parkLots[pNo] = {"alloted":False,"OName":None,"Vno":None,"ph":None,"rs":5}
  totalAmt+=5
  print(f"{pNo}z")

def viewExits():
  for i in exits:
    print(i)

def viewAlloted():
  for i in parkLots:
    if parkLots[i]['alloted']:
      print(parkLots[i])

def viewTotal():
  global totalAmt
  print(f"The total earned is :{totalAmt}")

print("Welcome to Parking Lot Management")

print("Available Slots (1 to 10)")

n = int(input("Enter the no of vehicle Bookings :"))
for k in range(n):
  sl = int(input("Enter Slot number :"))
  if checkEmpty(sl):
    vno = input("Enter vehicle number :")
    if vno in veh:
      print(f"{vno} is already alloted")
      continue
    else:
      name = input("Enter Owner name :")
      ph = input("Enter Phone number :")
      makeAllot(sl,name,vno,ph)
      veh.add(vno)
  else:
    print("Already Alloted !!")

ch = int(input("Enter the number of checkOuts :"))
for j in range(ch):
  cn = int(input("Enter the park slot number :"))
  makeCheckOut(cn)

print("The available slots are :")
for j in parkLots:
  if checkEmpty(j):
    print(parkLots[j])

print("The alloted slots are :")
viewAlloted()


print("Total Earned today :")
viewTotal()
```

```python
class Vehicle:
    def __init__(self, number, owner, phone):
        self.number = number
        self.owner = owner
        self.phone = phone


class ParkingSystem:
    def __init__(self):
        self.slots = {i: {"status": False, "vehicle": None, "fee": 5, "vnum": None} for i in range(1, 11)}
        self.vehicles_inside = set()
        self.history = {}
        self.total_money = 0
        self.his_id = 1

    def is_free(self, slot):
        return not self.slots[slot]["status"]

    def allot_slot(self, slot, vehicle):
        self.slots[slot]["vehicle"] = vehicle
        self.slots[slot]["status"] = True
        self.slots[slot]["vnum"] = vehicle.number
        self.vehicles_inside.add(vehicle.number)

        print(f"{vehicle.number} allotted to slot {slot} successfully")

    def checkout(self, slot):
        if self.is_free(slot):
            print("Slot already empty !!")
            return

        data = self.slots[slot]
        vehicle = data["vehicle"]

        self.history[self.his_id] = {"slot": slot, "vehicle": vehicle}
        self.his_id += 1

        self.vehicles_inside.remove(vehicle.number)
        self.total_money += data["fee"]

        self.slots[slot] = {"status": False, "vehicle": None, "fee": 5, "vnum": None}

        print(f"Slot {slot} is now empty")

    def show_allotted(self):
        print("Allotted Slots:")
        for s, info in self.slots.items():
            if info["status"]:
                print(f"Slot {s} allotted to vehicle : {info['vnum']}")

    def show_total(self):
        print("Total earned:", self.total_money)


print("Welcome to Parking Lot System")
print("Slots Available: 1 - 10")

parking = ParkingSystem()

bookings = int(input("Enter number of vehicle bookings: "))

for i in range(bookings):
    slot = int(input("Enter slot number: "))

    if slot not in parking.slots:
        print("Invalid slot number!")
        continue

    if parking.is_free(slot):
        vnum = input("Enter vehicle number: ")

        if vnum in parking.vehicles_inside:
            print("Vehicle already parked!")
            continue

        owner = input("Enter owner name: ")
        phone = input("Enter phone number: ")

        v = Vehicle(vnum, owner, phone)
        parking.allot_slot(slot, v)

    else:
        print("Slot already occupied!")

checkouts = int(input("Enter number of checkouts: "))

for i in range(checkouts):
    slot = int(input("Enter slot number to checkout: "))

    if slot in parking.slots:
        parking.checkout(slot)
    else:
        print("Invalid slot number!")

print()
parking.show_allotted()

print()
parking.show_total()
```

```python
class TaskManager:
    def __init__(self):
        self.task_list = {}
        self.task_id = 1

    def add_task(self):
        task = input("Enter your task: ")
        deadline = input("Enter deadline (dd/mm/yyyy): ")
        priority = int(input("Enter priority: "))
        tag = input("Enter task tag (Pending): ")

        self.task_list[self.task_id] = {
            "task": task,
            "deadline": deadline,
            "priority": priority,
            "status": tag
        }

        self.task_id += 1

    def update_status(self, tid):
        if tid not in self.task_list:
            print("Task ID not found !!")
            return

        new_status = input("Enter the status of the task: ")
        self.task_list[tid]["status"] = new_status

    def filter_tasks(self, status):
        print(f"Tasks filtered according to {status}")
        for i, data in self.task_list.items():
            if data["status"] == status:
                print(f"TaskID: {i}, Task: {data['task']}, Deadline: {data['deadline']}")

    def sort_priority(self):
        sorted_tasks = sorted(self.task_list.items(),
                              key=lambda x: x[1]["priority"])

        for tid, task in sorted_tasks:
            print(tid, task)

    def show_tasks(self):
        for i, data in self.task_list.items():
            print(f"TaskID: {i}, Task: {data['task']}, Deadline: {data['deadline']}")


tm = TaskManager()

n = int(input("Enter the number of tasks to enter: "))
for i in range(n):
    tm.add_task()

print("All the tasks are:")
tm.show_tasks()

n2 = int(input("Enter the number of tasks to change the status: "))
for j in range(n2):
    tid = int(input("Enter the task ID: "))
    tm.update_status(tid)

print("View tasks by tag:")
tag = input("Enter tag to sort: ")
tm.filter_tasks(tag)

print("View Task based on priority:")
tm.sort_priority()
```

```python
class Store:
  def __init__(self):
    self.products = {}
    self.pid = 1

  def addProduct(self):
    prod = input("Enter your Product name :")
    stock = int(input("Enter stock number :"))
    pr = float(input("Enter its price:"))
    disc = int(input("Enter the discount :"))
    discPrice = pr - (disc/100)*pr
    self.products[self.pid] = {'Product':prod,'stock':stock,'OPrice':pr,"Discount":disc,"DiscPr":discPrice}
    self.pid+=1

  def changeDiscount(self):
    id = int(input("Enter the id of the product :"))
    if id not in self.products:
      print("Product not found !!")
      return
    disc = int(input("Enter the discount :"))
    pr = self.products[id]['OPrice']
    discPrice = pr - (disc/100)*pr
    self.products[id]['Discount'] = disc
    self.products[id]['DiscPr'] = discPrice
    print("Discount updated successfully !!")

  def viewProduct(self):
    for i in self.products:
      p = self.products[i]
      print(f"ProductID : {i}, Product : {p['Product']} ,stock : {p['stock']} ,Price : {p['OPrice']} ,Discount Price : {p['DiscPr']}")

  def makePurchase(self):
    print("All product :")
    self.viewProduct()
    cart = []
    n = int(input("Enter the no of item to purchase :"))
    total = 0
    for i in range(n):
      pd = int(input("Enter the product ID :"))
      if pd not in self.products:
        print("Invalid Product ID")
        continue
      quan = int(input("Enter the number of quantity :"))
      if quan > self.products[pd]['stock']:
        print("Quantity is greater then available stock")
        continue
      price = self.products[pd]['DiscPr']
      cost = price * quan
      self.products[pd]['stock'] -= quan
      cart.append((pd,quan,cost))
      total += cost
    print("Cart details :")
    for c in cart:
      print("ProductID :",c[0],"Quantity :",c[1],"Cost :",c[2])
    print("Total Bill :",total)

  def sortByPrice(self):
    s = sorted(self.products.items(), key=lambda x: x[1]['DiscPr'])
    for pid,p in s:
      print(pid,p)


s1 = Store()

n = int(input("Enter the number of products to add :"))
for i in range(n):
  s1.addProduct()

print("All the products are :")
s1.viewProduct()

n2 = int(input("Enter number of products to change discount :"))
for j in range(n2):
  s1.changeDiscount()

print("Products after discount update :")
s1.viewProduct()

print("Make Purchase :")
s1.makePurchase()

print("Products sorted by price :")
s1.sortByPrice()
```

3.Banking System with Transactions
Simulate real-time transactions between bank accounts.
Handle errors like overdraft, transaction timeout, incorrect account numbers.

```python
class InvalidAccountNumber(Exception):
  def __init__(self):
    print("Invalid Account Number")

class AccountAlreadyExist(Exception):
  def __init__(self):
    print("Account already exist")

class NoSuchAccount(Exception):
  def __init__(self):
    print("No Such Account exist")

class OverDraftError(Exception):
  def __init__(self):
    print("Insufficient Balance")

class BankingSystem:
  def __init__(self):
    self.accounts = {}

  def createAccount(self):
    try:
      acc = input("Enter Account Number :")
      if len(acc) < 5:
        raise InvalidAccountNumber()
        return
      if acc in self.accounts:
        raise AccountAlreadyExist()
        return
      bal = float(input("Enter Initial Balance :"))
      self.accounts[acc] = bal
      print("Account Created Successfully")
    except InvalidAccountNumber as e:
      print(e)
    except AccountAlreadyExist as e:
      print(e)
    except Exception as e:
      print(e)

  def depositMoney(self):
    try:
      acc = input("Enter Account Number :")
      if acc not in self.accounts:
        raise NoSuchAccount()
        return
      amt = float(input("Enter Amount to Deposit :"))
      self.accounts[acc] += amt
      print("Deposit Successful")
    except NoSuchAccount as e:
      print(e)
    except Exception as e:
      print(e)

  def transferMoney(self):
    try:
      acc1 = input("Enter Sender Account Number :")
      acc2 = input("Enter Receiver Account Number :")
      if acc1 not in self.accounts or acc2 not in self.accounts:
        raise NoSuchAccount()
        return
      amt = float(input("Enter Amount to Transfer :"))
      if self.accounts[acc1] < amt:
        raise OverDraftError()
        return
      self.accounts[acc1] -= amt
      self.accounts[acc2] += amt
      print("Transaction Successful")
    except NoSuchAccount as e:
      print(e)
    except OverDraftError as e:
      print(e)
    except Exception as e:
      print(e)

  def checkBalance(self):
    try:
      acc = input("Enter Account Number :")
      if acc not in self.accounts:
        raise NoSuchAccount()
        return
      print(f"Balance of {acc} is {self.accounts[acc]}")
    except NoSuchAccount as e:
      print(e)
    except Exception as e:
      print(e)

bs = BankingSystem()

while True:
  print("1 for create account")
  print("2 for deposit money")
  print("3 for transfer money")
  print("4 for check balance")
  print("5 for exit")

  try:
    choice = int(input("Enter your choice : "))
    if choice == 1:
      bs.createAccount()
    elif choice == 2:
      bs.depositMoney()
    elif choice == 3:
      bs.transferMoney()
    elif choice == 4:
      bs.checkBalance()
    elif choice == 5:
      break
    else:
      print("Invalid Choice !")
  except:
    print("Invalid Choice Type")

print("Thank you")
```

4.E-Commerce Order Management
Manage orders, returns, and refunds.
Handle cases like invalid coupon code, out-of-stock errors, invalid payment
methods.

```python
class InvalidCoupon(Exception):
  def __init__(self):
    print("Invalid Coupon Code")

class OutOfStock(Exception):
  def __init__(self):
    print("Product Out Of Stock")

class InvalidPaymentMethod(Exception):
  def __init__(self):
    print("Invalid Payment Method")

class NoSuchOrder(Exception):
  def __init__(self):
    print("No Such Order Exist")

class Ecommerce:
  def __init__(self):
    self.products = {"laptop":5,"phone":10,"headphone":7}
    self.orders = {}
    self.orderId = 1
    self.coupon = ['SAVE','FREE','OFFER']

  def placeOrder(self):
    try:
      p = input("Enter product name :")
      if p not in self.products:
        raise OutOfStock()
        return
      if self.products[p] == 0:
        raise OutOfStock()
        return
      coupon = input("Enter coupon code (or press enter) :")
      if coupon not in self.coupon:
        raise InvalidCoupon()
        return
      pay = input("Enter payment method (card/upi/cod) :")
      if pay not in ["card","upi","cod"]:
        raise InvalidPaymentMethod()
        return
      oid = self.orderId
      self.orders[oid] = p
      self.products[p] -= 1
      self.orderId += 1
      print(f"Order placed successfully with Order ID {oid}")
    except InvalidCoupon as e:
      print(e)
    except OutOfStock as e:
      print(e)
    except InvalidPaymentMethod as e:
      print(e)
    except Exception as e:
      print(e)

  def returnOrder(self):
    try:
      oid = int(input("Enter order id :"))
      if oid not in self.orders:
        raise NoSuchOrder()
        return
      p = self.orders[oid]
      self.products[p] += 1
      del self.orders[oid]
      print("Order returned successfully")
    except NoSuchOrder as e:
      print(e)
    except Exception as e:
      print(e)

  def refund(self):
    try:
      oid = int(input("Enter order id :"))
      if oid not in self.orders:
        raise NoSuchOrder()
        return
      del self.orders[oid]
      print("Refund processed successfully")
    except NoSuchOrder as e:
      print(e)
    except Exception as e:
      print(e)

  def viewOrders(self):
    if len(self.orders) == 0:
      print("No Orders")
    else:
      for i in self.orders:
        print(f"Order {i} : {self.orders[i]}")

ec = Ecommerce()

while True:
  print("1 for place order")
  print("2 for return order")
  print("3 for refund")
  print("4 for view orders")
  print("5 for exit")

  try:
    ch = int(input("Enter your choice :"))
    if ch == 1:
      ec.placeOrder()
    elif ch == 2:
      ec.returnOrder()
    elif ch == 3:
      ec.refund()
    elif ch == 4:
      ec.viewOrders()
    elif ch == 5:
      break
    else:
      print("Invalid Choice")
  except:
    print("Invalid Choice Type")

print("Thank you")
```

```python
class OutOfStockError(Exception):
  def __init__(self):
    print("Product is out of stock")

class InvalidProductIDError(Exception):
  def __init__(self):
    print("Invalid product id")

class ProductNotFoundError(Exception):
  def __init__(self):
    print("Product not found")

class Inventory:
  def __init__(self):
    self.products = {1:["pen",10],2:["book",5],3:["bag",2]}

  def viewProducts(self):
    if len(self.products) == 0:
      print("No products")
    else:
      for i in self.products:
        print(i,self.products[i][0],self.products[i][1])

  def buyProduct(self):
    try:
      pid = int(input("Enter product id :"))
      if pid not in self.products:
        raise InvalidProductIDError()
        return
      qty = int(input("Enter quantity :"))
      if self.products[pid][1] == 0 or qty > self.products[pid][1]:
        raise OutOfStockError()
        return
      self.products[pid][1] -= qty
      print("Product purchased")
    except InvalidProductIDError as e:
      print(e)
    except OutOfStockError as e:
      print(e)
    except Exception as e:
      print(e)

  def addStock(self):
    try:
      pid = int(input("Enter product id :"))
      if pid not in self.products:
        raise ProductNotFoundError()
        return
      qty = int(input("Enter quantity to add :"))
      self.products[pid][1] += qty
      print("Stock added")
    except ProductNotFoundError as e:
      print(e)
    except Exception as e:
      print(e)

inv = Inventory()

while True:
  print("1 view products")
  print("2 buy product")
  print("3 add stock")
  print("4 exit")

  try:
    ch = int(input("Enter choice :"))
    if ch == 1:
      inv.viewProducts()
    elif ch == 2:
      inv.buyProduct()
    elif ch == 3:
      inv.addStock()
    elif ch == 4:
      break
    else:
      print("Invalid choice")
  except:
    print("Invalid input")

print("Program finished")
```

Handle exceptions such as seat not available, invalid passenger details, payment
failure.

```python
class SeatNotAvailable(Exception):
  def __init__(self):
    print("Seat not available")

class InvalidPassengerDetails(Exception):
  def __init__(self):
    print("Invalid passenger details")

class PaymentFailure(Exception):
  def __init__(self):
    print("Payment failed")

class NoSuchBooking(Exception):
  def __init__(self):
    print("No such booking")

class FlightSystem:
  def __init__(self):
    self.flights = {"AI101":3,"AI202":2,"AI303":4}
    self.bookings = {}
    self.bookId = 1

  def searchFlights(self):
    for i in self.flights:
      print(i,"Seats:",self.flights[i])

  def bookFlight(self):
    try:
      name = input("Enter passenger name :")
      if name == "":
        raise InvalidPassengerDetails()
        return
      f = input("Enter flight number :")
      if f not in self.flights:
        raise SeatNotAvailable()
        return
      if self.flights[f] == 0:
        raise SeatNotAvailable()
        return
      pay = input("Enter payment status (success/fail) :")
      if pay != "success":
        raise PaymentFailure()
        return
      bid = self.bookId
      self.bookings[bid] = [name,f]
      self.flights[f] -= 1
      self.bookId += 1
      print("Flight booked with booking id",bid)
    except SeatNotAvailable as e:
      print(e)
    except InvalidPassengerDetails as e:
      print(e)
    except PaymentFailure as e:
      print(e)
    except Exception as e:
      print(e)

  def cancelBooking(self):
    try:
      bid = int(input("Enter booking id :"))
      if bid not in self.bookings:
        raise NoSuchBooking()
        return
      f = self.bookings[bid][1]
      self.flights[f] += 1
      del self.bookings[bid]
      print("Booking cancelled")
    except NoSuchBooking as e:
      print(e)
    except Exception as e:
      print(e)

  def viewBookings(self):
    if len(self.bookings) == 0:
      print("No bookings")
    else:
      for i in self.bookings:
        print(i,self.bookings[i][0],self.bookings[i][1])

fs = FlightSystem()

while True:
  print("1 search flights")
  print("2 book flight")
  print("3 cancel booking")
  print("4 view bookings")
  print("5 exit")

  try:
    ch = int(input("Enter choice :"))
    if ch == 1:
      fs.searchFlights()
    elif ch == 2:
      fs.bookFlight()
    elif ch == 3:
      fs.cancelBooking()
    elif ch == 4:
      fs.viewBookings()
    elif ch == 5:
      break
    else:
      print("Invalid choice")
  except:
    print("Invalid input")

print("Thank you")
```
